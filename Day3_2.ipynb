```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
X = iris.data
X
```

```python
y =iris.target
y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y.shape
```

```python
y
```

```python
pip install tensorflow
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_object = MinMaxScaler()
```

```python
scalar_object.fit(X_train)
```

```python
scaled_X_train = scalar_object.transform(X_train)
```

```python
scaled_X_test = scalar_object.transform(X_test)
```

```python
X_train.max()
```

```python
scaled_X_train.max()
```

```python
X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
scaled_X_test
```

```python
model.predict(scaled_X_test)
```

```python
model.metrics_names
```

```python
model.evaluate(x=scaled_X_test,y=y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions = (model.predict(x=scaled_X_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
y_test.argmax(axis=1)
```

```python
confusion_matrix(y_test.argmax(axis=1),predictions.argmax(axis=1))
```

```python
print(classification_report(y_test.argmax(axis=1),predictions.argmax(axis=1)))
```
