```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
pattern1 =[{'LOWER': 'solarpower'}]
pattern2 = [{'LOWER': 'solar'}, {'LOWER': 'power'}]
pattern3 = [{'LOWER': 'Solarpower'}]
pattern4 = [{'LOWER': 'solar'}, {'IS_PUNCT': True}, {'LOWER': 'power'}]
```

```python
doc = nlp(u"The solarpower industry continues to grow as demand for solar energy increases.Solarpower cars are gaining popularity.")
```

```python
found_matches = matcher(doc)
print(found_matches)
```
