```python
#sentence segregation
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("This is the first sentence. This is another sentence. This is the last sentence.")
```

```python
doc_sents = list(doc.sents)
print(doc_sents[1])
```

```python
doc2 = nlp(u"This is the first sentence. This is another sentence. This is the last sentence.")
for token in doc2:
  print(token.is_sent_start, ''+token.text)
```

```python
doc=nlp(u'"Management is doing the right things;leadership is doing the right things."-Peter Drucker')
for sent in doc.sents:
    print(sent)
    print('\n')
```

```python
from spacy.language import Language
@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
  for token in doc[:-1]:
    if token.text==';':
      if token.i + 1 < len(doc):
        doc[token.i + 1].is_sent_start = True
  return doc
```

```python
if 'set_custom_boundaries' in nlp.pipe_names:
    nlp.remove_pipe('set_custom_boundaries')
nlp.add_pipe("set_custom_boundaries", before='parser')
print(nlp.pipe_names)
```

```python
doc4=nlp(u'''"Management is doing things right;
leadership is doing the right things;"
-Peter Drucker''', disable=['set_custom_boundaries'])
for sent in doc4.sents:
    print(sent)
    print('\n')
```
