```python
#to check whether there is full image
full = cv2.imread('/content/bricks.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)

#template match
face = cv2.imread('/content/sammy_face.jpg') # Read as color image
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

# Get dimensions of the face image for rectangle drawing
height, width,channels= face.shape

#6 methods comparison in a list
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
  full_copy = full.copy()
  method = eval(m)
  res = cv2.matchTemplate(full_copy, face, method)
  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
  #notices the color of last 2 left side img
  if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc
  else:
    top_left = max_loc

    #assign to the bottom right of the rectangle
  bottom_right = (top_left[0] + width, top_left[1] + height)

  #draw red rectangle
  cv2.rectangle(full_copy, top_left, bottom_right, (255, 0, 0), 3)

  #plot the images
  plt.subplot(121)
  plt.imshow(res)
  plt.title('Matching Result')
  plt.subplot(122)
  plt.imshow(full_copy)
```

```python
plt.subplot(122)
plt.imshow(full_copy)
plt.title('Result of Template Matching')
```
