```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img = np.zeros((500,500,3),dtype='uint8')
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img,(384,0),(510,128),(0,255,0),thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(511,511), color=(102,255,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
#Text
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello',org=(10,500), fontFace=font, fontScale=4, color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```

```python
#polygons
```

```python
blank_img = np.zeros((500,500,3),dtype='uint8')
```

```python
vertices = np.array([[10,50],[20,30],[20,50],[30,60]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
blank_img =cv2.imread("dog_backpack.jpg",0)
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```
