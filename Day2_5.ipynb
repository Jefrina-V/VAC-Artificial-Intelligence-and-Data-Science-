```python
def show_ents(doc):
  if doc.ents:
    for ent in doc.ents:
      print(ent.text+'_'+ent.label_+'_'+str(spacy.explain(ent.label_)))
  else:
        print('No named entities found')
```

```python
doc = nlp(u'May I go to Washington, DC next May to see the Washington Monuments?')
show_ents(doc)
```

```python
doc = nlp(u'Can I please have 500 dollars  from you to buy some Microsoft stock?')
for ent in doc.ents:
  print(ent.text, ent.start, ent.end, ent.start_char,ent.end_char,ent.label_)
```

```python
doc = nlp(u"Tesla to build a U.K. factory factory for $6 million")
show_ents(doc)
```

```python
doc = nlp(u"Tesla to build a U.K. factory factory for $6 million")
ORG = doc.vocab.strings[u'ORG']
new_ent = spacy.tokens.Span(doc,0,1,label=ORG)
doc.ents = list(doc.ents) + [new_ent]
show_ents(doc)
```

```python
doc = nlp(u"Our community plans to introduce a new vaccum cleaner"
          u"If successful, the vacuum cleaner will be our first product.")
show_ents(doc)
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
doc = nlp(u"Our Company plans to introduce a new vaccum cleaner.")
Phrase_list = ['vaccum cleaner','vaccum cleaner']
phrase_patterns = [nlp(text) for text in Phrase_list]
matcher.add('newproduct',None,*phrase_patterns)
matches = matcher(doc)
matches
```
